```python
np.random.seed(42)
df = pd.read_excel('clusters.xlsx')
```

```python
df.head(5)
```

```python
wcss = []
for i in range(2, 10):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(df)
    wcss.append(round(kmeans.inertia_, 2))
print(3)
```
