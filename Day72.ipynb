```python
# how does it look like ??? 
df = pd.read_csv('pca.csv')
df.head(5)
```

```python
# find value : values() is an inbuilt method in Python programming language that returns a view object 
# The view object contains the values of the dictionary, as a list.
data = df.values
data
```

```python
# hmm let's use standardscaler to standardize this data
scaler = StandardScaler()
#  training data to scale the data and also learn the scaling parameters
data_std = scaler.fit_transform(data)
```

```python
pca = PCA(n_components=3)
data_pca = pca.fit_transform(data_std)
```

```python
# we get ... under the table form
results = pd.DataFrame(
    data={'explained_variance_ratio': pca.explained_variance_ratio_}
)
results['cumulative'] = results['explained_variance_ratio'].cumsum()
results['component'] = results.index + 1
print(results)
```
