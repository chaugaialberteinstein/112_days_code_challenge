Exercise 12

The df DataFrame is given below:


                   currency
    0            [PLN, USD]
    1  [EUR, USD, PLN, CAD]
    2                 [GBP]
    3       [JPY, CZK, HUF]
    4                    []


Assign a new column 'number' that takes the number of items in the currency column.

In response, print df DataFrame to the console.


Expected result:


                   currency  number
    0            [PLN, USD]       2
    1  [EUR, USD, PLN, CAD]       4
    2                 [GBP]       1
    3       [JPY, CZK, HUF]       3
    4                    []       0




```python
import pandas as pd

df = pd.DataFrame(data ={ 'currency' : [ ['PLN', 'USD'],
                                 ['EUR', 'USD', 'PLN', 'CAD'] ,
                                 ['GBP'],
                                 ['JPY', 'CZK', 'HUF'],
                                 [] ]})
df
```

```python
df['number'] = df['currency'].apply(len)
df
```
