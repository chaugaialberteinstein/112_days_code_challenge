The df DataFrame is given below:


       investments
    0  100_000_000
    1      100_000
    2   30_000_000
    3  100_500_000


Convert column investments to int type and print result to the console.


Expected result:


       investments
    0    100000000
    1       100000
    2     30000000
    3    100500000




```python
import pandas as pd
df = pd.DataFrame({'investment':['100_000_000','100_000','30_000_000','100_500_000']})
# keychecked
df['investment'] = df['investment'].str.replace('_', '').astype(int)                      
df
```
